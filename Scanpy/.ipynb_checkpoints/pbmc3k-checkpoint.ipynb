This is a recreation of the Scanpy [tutorial](https://scanpy-tutorials.readthedocs.io/en/latest/pbmc3k.html) that considers the PBMC3K dataset. Essentially, it recreates the original Seurat tutorial within the Scanpy framework.

First we import the relevant packages: numpy, pandas, and scanpy.

```python
import numpy as np
import pandas as pd
import scanpy as sc
```

First, we initialise some settings for scanpy.

```python
sc.settings.verbosity = 3 # Possible values: (0) errors, (1) warnings, (2) info, (3) hints
sc.logging.print_versions()
sc.settings.set_figure_params(dpi = 80, facecolor='white')
```

```python
results_file = 'pbmc3k.h5ad'
```

We now read in the data for the PBMC3K dataset.

```python
adata = sc.read_10x_mtx('../Data/filtered_gene_bc_matrices/hg19/', # Directory with relevant .mtx file
        var_names = 'gene_symbols', # Gene symbols will be used as variable names
        cache=True) # By writing a cache file, it makes subsequent reading faster.
adata.var_names_make_unique() # If var_names = 'gene_ids', when this step isn't necessary
adata
```

Let's start pre-processing the data. We first look at the top 20 genes with the highest fraction of counts in each single cell (across all cells).

```python
sc.pl.highest_expr_genes(adata, n_top=20, )
```

We filter out cells that express fewer than 200 genes and filter out genes that are detected in fewer than 3 cells.

```python
sc.pp.filter_cells(adata, min_genes=200)
sc.pp.filter_genes(adata, min_cells=3)
```

We calculate mitochondrial gene expression so that we can filter these cells out.

```python
adata.var['mt'] = adata.var_names.str.startswith('MT-') # Dictionary with 'mt' key corresponding to genes that start with MT-
sc.pp.calculate_qc_metrics(adata, qc_vars=['mt'], percent_top=None, log1p=False, inplace=True) # To be honest, I don't know what these options do.
```

We produce the same violin plots from the Seurat tutorial:
1. The number of genes expressed in the count matrix.
2. The total counts per cell.
3. The percentage of counts in mitochondrial genes.

```python
sc.pl.violin(adata, ['n_genes_by_counts', 'total_counts', 'pct_counts_mt'], jitter=0.4, multi_panel=True)
sc.pl.scatter(adata, x='total_counts', y='pct_counts_mt')
sc.pl.scatter(adata, x='total_counts', y='n_genes_by_counts')
```

We filter out cells that have an abnormally high gene count, because those could be doublets/multiplets, as well as filtering out cells with more than 5% of their gene expression consisting of mitochondrial genes.

```python
adata = adata[adata.obs.n_genes_by_counts < 2500, :] # Remove cells with abnormally high gene expression counts
adata = adata[adata.obs.pct_counts_mt < 5, :] # Remove cells with abnormally high mitochondrial gene expression
```

In order to compare counts across cells, we need to normalise the data. We normalise by the total count (library-size correct), so that the data matrix is normalised to 10, 000 reads per cell. We then log-transform the data, to allow us to compare reads of varying orders of magnitude.

```python
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
```

We now identify the highly-variable genes.

```python
sc.pp.highly_variable_genes(adata, min_mean=0.0125, max_mean=3, min_disp=0.5) # No idea what these parameters mean.
```

```python
sc.pl.highly_variable_genes(adata)
```

Set the raw attribute of the data to the normalised and log-transformed data, so that we can use this dataset for differential testing and gene expression visualisation later.

```python
adata.raw = adata
```

Filter out the genes that do not vary much across cells.

```python
adata = adata[:, adata.var.highly_variable] 
```

```python
sc.pp.regress_out(adata, ['total_counts', 'pct_counts_mt']) 
```

Scale each gene to unit variance, clipping genes with standard deviations greater than ten.

```python
sc.pp.scale(adata, max_value=10)
```

We now reduce the dimensionality of the dataset by first running a Principal Component Analysis (PCA), which reveals the main axes of variation and denoises the data (apparently).

```python
sc.tl.pca(adata, svd_solver='arpack')
```

To give an idea of the spread, let's produce a scatter plot of PC1 vs PC2. 

```python
sc.pl.pca(adata, color='CST3')
```

We will now inspect the contribution of each PC to the total variance in the data, using an elbow plot. This gives us an idea of the dimensionality of the dataset and thus how many PCs to incorporate when we cluster the data. It seems that we only need a rough estimate of the number of PCs.

```python
sc.pl.pca_variance_ratio(adata, log=True)
```

Save the results thus far.

```python
adata.write(results_file)
```

```python
adata
```

We are now ready to start clustering the data. First, we use the results from the PCA to construct the kNN graph. The graph is then embedded in 2D using a [UMAP](https://arxiv.org/abs/1802.03426) representation.

```python
sc.pp.neighbors(adata, n_neighbors=10, n_pcs=40)
```

The reason we use UMAP over tSNE is that it preserves global connectivity better, which makes it more useful for inferring trajectories. If we still see disconnected clusters, we can usually attempt to fix these by running the following commands: 
- tl.paga(adata)
- pl.paga(adata, plot=False) # Remove plot=False if we want to see the coarse-grained graph
- tl.umap(adata, init_pos='paga')

Compute the UMAP.

```python
sc.tl.umap(adata)
```

```python
sc.pl.umap(adata, color=['CST3', 'NKG7', 'PPBP'])
sc.pl.umap(adata, color=['CST3', 'NKG7', 'PPBP'], use_raw=False) # If we don't want to use the raw data, but the normalised/scaled data
```

We now cluster based on the Leiden method (community detection based on optimizing modularity); see [Traag et al.](https://arxiv.org/abs/1810.08473) for further details. This method clusters the kNN graph that we constructed in the previous steps.

```python
sc.tl.leiden(adata)
```

Plot the clusters.

```python
sc.pl.umap(adata, color=['leiden', 'CST3', 'NKG7'])
```

Save the results.

```python
adata.write(results_file)
```
